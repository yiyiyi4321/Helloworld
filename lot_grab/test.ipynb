```python
import pandas as pd
import numpy as np

def lotobj(preds, dtrain):
    w = dtrain.get_label()
    y = 1.0 / (1.0 + np.exp(-preds))
    print(w, y)
    grad = np.zeros((y.shape[0], 4))
    for i in range(y.shape[0]):
        grad[i,0] = y[i,0]*(1.0-y[i,0])*(y[i,1]*w[i,1] + y[i,2]*w[i,2] + y[i,3]*w[i,3])
        grad[i,1] = y[i,0]*(y[i,1]*(1.0-y[i,1])*w[i,1] - y[i,1]*y[i,2]*w[i,2] - y[i,1]*y[i,3]*w[i,3])
        grad[i,2] = y[i,0]*(y[i,2]*(1.0-y[i,2])*w[i,2] - y[i,2]*y[i,1]*w[i,1] - y[i,2]*y[i,3]*w[i,3])
        grad[i,3] = y[i,0]*(y[i,3]*(1.0-y[i,3])*w[i,3] - y[i,3]*y[i,1]*w[i,1] - y[i,3]*y[i,2]*w[i,2])
    hess = np.zeros((y.shape[0], 4))
    for i in range(y.shape[0]):
        hess[i,0] = y[i,0]*(1.0-y[i,0])*(1.0-2.0*y[i,0])*(y[i,1]*w[i,1] + y[i,2]*w[i,2] + y[i,3]*w[i,3])
        hess[i,1] = y[i,0]*(y[i,1]*(1.0-y[i,1])*(1.0-2.0*y[i,1])*w[i,1] - y[i,1]*(1.0-y[i,1])*y[i,2]*w[i,2] - y[i,1]*(1.0-y[i,1])*y[i,3]*w[i,3])
        hess[i,2] = y[i,0]*(y[i,2]*(1.0-y[i,2])*(1.0-2.0*y[i,2])*w[i,2] - y[i,2]*(1.0-y[i,2])*y[i,1]*w[i,1] - y[i,2]*(1.0-y[i,2])*y[i,3]*w[i,3])
        hess[i,3] = y[i,0]*(y[i,3]*(1.0-y[i,3])*(1.0-2.0*y[i,3])*w[i,3] - y[i,3]*(1.0-y[i,3])*y[i,1]*w[i,1] - y[i,3]*(1.0-y[i,3])*y[i,2]*w[i,2])
    return grad, hess

def lotscore(preds, dtrain):
    w = dtrain.get_label()
    y = 1.0 / (1.0 + np.exp(-preds))
    res = np.zeros(y.shape[0])
    for i in range(y.shape[0]):
        res[i] = y[i,0]*(y[i,1]*w[i,1] + y[i,2]*w[i,2] + y[i,3]*w[i,3])
    return ('lotscore', 0-res)


yingchao_data = pd.read_csv('lot_500_football_yingchao.csv')
#xijia_data = pd.read_csv('lot_500_football.csv')
#yingchao_data = yingchao_data.append(xijia_data)
#yingchao_data = xijia_data.append(yingchao_data)

yingchao_data['label0'] = 0
yingchao_data['label1'] = yingchao_data['f082']
yingchao_data['label2'] = yingchao_data['f083']
yingchao_data['label3'] = yingchao_data['f084']
print(yingchao_data.shape)
print(yingchao_data.iloc[:2000, -142:-1])
import xgboost as xgb

dtrain = xgb.DMatrix(yingchao_data.iloc[886:, -148:-67].values, label=yingchao_data.iloc[886:, -4:].values)
dtest = xgb.DMatrix(yingchao_data.iloc[:886, -148:-67].values, label=yingchao_data.iloc[:886, -4:].values)

param = {'max_depth': 4, 'eta': 0.001}
evallist = [(dtest, 'eval'), (dtrain, 'train')]

num_round = 888
bst = xgb.train(param, dtrain, num_round, evallist, obj=lotobj, feval=lotscore)

ypreds_1 = bst.predict(dtest)

dtrain = xgb.DMatrix(yingchao_data.iloc[886:, -148:-4].values, label=yingchao_data.iloc[886:, -4:].values)
dtest = xgb.DMatrix(yingchao_data.iloc[:886, -148:-4].values, label=yingchao_data.iloc[:886, -4:].values)

param = {'max_depth': 4, 'eta': 0.001}
evallist = [(dtest, 'eval'), (dtrain, 'train')]

num_round = 888
bst = xgb.train(param, dtrain, num_round, evallist, obj=lotobj, feval=lotscore)

ypreds_2 = bst.predict(dtest)


labels = dtest.get_label()
```
