```python
import math
from queue import PriorityQueue as PQueue
import re
import urllib.request, zlib
import time
import sys
import string
from selenium import webdriver

def get_html(url):
    browser = webdriver.Firefox()
    content = ""
    fail_count = 0
    while(len(content) < 10000 and fail_count < 5):
        try:
            browser.get(url)
            content = browser.page_source
            if len(content) < 10000:
                fail_count += 1
        except:
            fail_count += 1
    print(url, fail_count, len(content))
    browser.quit()
    return content

def get_data_set(shuju_html, str_id):
    reg = []
    reg.append('<div class="M_title">\n\
      <h4>赛前联赛积分排名</h4>\n\
    </div>\n\
    <div class="M_sub_title">\n\
    <div class="team_name">(.*)<span>\[(.*)\]</span></div>\n\
    <div class="team_name">(.*)<span>\[(.*)\]</span></div>')
    reg.append('<p class="game_time">比赛时间(.*)</p>\n\
            <p class="odds_hd_bf"><strong>(\d*):(\d*)</strong></p>')
    reg.append('<td class="td_one">总成绩</td><td>([0-9]\d*)</td><td>([0-9]\d*)</td><td>([0-9]\d*)</td><td>([0-9]\d*)</td><td>([0-9]\d*)</td><td>([0-9]\d*)</td><td>([-|0-9]\d*)</td><td><span class="ying">([0-9]\d*)</span></td><td>([0-9]\d*)</td><td>([0-9]\d*)%</td>')
    reg.append('<td class="td_one">主场</td><td>([0-9]\d*)</td><td>([0-9]\d*)</td><td>([0-9]\d*)</td><td>([0-9]\d*)</td><td>([0-9]\d*)</td><td>([0-9]\d*)</td><td>([-|0-9]\d*)</td><td><span class="ying">([0-9]\d*)</span></td><td>([0-9]\d*)</td><td>([0-9]\d*)%</td>')
    reg.append('<td class="td_one">客场</td><td>([0-9]\d*)</td><td>([0-9]\d*)</td><td>([0-9]\d*)</td><td>([0-9]\d*)</td><td>([0-9]\d*)</td><td>([0-9]\d*)</td><td>([-|0-9]\d*)</td><td><span class="ying">([0-9]\d*)</span></td><td>([0-9]\d*)</td><td>([0-9]\d*)%</td>')
    reg.append('<span class="his_info">双方近<span class="fb2">([0-9]\d*)</span>次交战，.*<span class="f16"><em class="red">([0-9]\d*)胜</em><em class="green">([0-9]\d*)平</em><em class="blue">([0-9]\d*)负</em></span>，进([0-9]\d*)球，失([0-9]\d*)球，大球([0-9]\d*)次，小球([0-9]\d*)次</span>')
    reg.append('近([0-9]\d*)场战绩<span class="mar_left20"><span class="ying">([0-9]\d*)胜</span><span class="ping">([0-9]\d*)平</span><span class="shu">([0-9]\d*)负</span></span><span class="mar_left20">进<span class="ying">([0-9]\d*)球</span>失<span class="shu">([0-9]\d*)球</span></span></p>')
    
    line = str_id
    count = 0
    
    page = re.compile(reg[0])
    rs = re.findall(page,shuju_html)
    if len(rs) == 1:
        count += len(rs)
        line += "," + rs[0][0] + "," + rs[0][2]
    
    page = re.compile(reg[1])
    rs = re.findall(page,shuju_html)
    if len(rs) == 1:
        count += len(rs)
        line += "," + rs[0][0] + "," + rs[0][1] + "," + rs[0][2]
    
    for i in range(2, 5):
        page = re.compile(reg[i])
        rs = re.findall(page,shuju_html)
        count += len(rs)
        for r in rs:
            for j in range(1, len(r)-2):
                line += "," + str(int(r[j])/int(r[0]))
            line += "," + str(math.log(int(r[-2])))
            line += "," + str(int(r[-1])/100)
            
    page = re.compile(reg[5])
    rs = re.findall(page,shuju_html)
    if len(rs) == 1:
        count += len(rs)
        for j in range(1, len(rs[0])):
            line += "," + str(int(rs[0][j])/int(rs[0][0]))
        
    page = re.compile(reg[6])
    rs = re.findall(page,shuju_html)
    count += len(rs)
    for r in rs:
        for j in range(1, len(r)):
            line += "," + str(int(r[j])/int(r[0]))
    
    url = 'http://odds.500.com/fenxi/ouzhi-'+str_id+'.shtml'
    ouzhi_html = get_html(url)
    com = ['威廉希尔', '澳门', '立博']
    for c in com:
        reg = c + '[\s\S]*?> (\d\d*\.\d\d*)</td>[\s\S]*?>(\d\d*\.\d\d*)</td>[\s\S]*?> (\d\d*\.\d\d*)</td>\
[\s\S]*?>(\d\d*\.\d\d*)</td>[\s\S]*?>(\d\d*\.\d\d*)</td>[\s\S]*?>(\d\d*\.\d\d*)</td>\
[\s\S]*?>(\d\d*\.\d\d*)%</td>[\s\S]*?>(\d\d*\.\d\d*)%</td>[\s\S]*?>(\d\d*\.\d\d*)%</td>\
[\s\S]*?>(\d\d*\.\d\d*)%</td>[\s\S]*?>(\d\d*\.\d\d*)%</td>[\s\S]*?>(\d\d*\.\d\d*)%</td>\
[\s\S]*?>(\d\d*\.\d\d*)%</td>[\s\S]*?>(\d\d*\.\d\d*)%</td>\
[\s\S]*?>(\d\d*\.\d\d*)</td>[\s\S]*?>(\d\d*\.\d\d*)</td>[\s\S]*?>(\d\d*\.\d\d*)</td>\
[\s\S]*?>(\d\d*\.\d\d*)</td>[\s\S]*?>(\d\d*\.\d\d*)</td>[\s\S]*?>(\d\d*\.\d\d*)</td>'
        page = re.compile(reg)
        rs = re.findall(page,ouzhi_html)
        if len(rs) > 0:
            count += len(rs)
            for j in range(0, len(rs[0])):
                line += "," + rs[0][j]
    if count == 16:
        with open('lot_500_football.csv', 'a') as f:
            f.write(line+'\n')
    else:
        with open('lot_500_fail.csv', 'a') as f:
            f.write(str_id+' '+str(count)+'\n')
    return line

def bfs_grab():
    pq = PQueue()
    id_set = set()
    for id in range(573094, 573104):
        pq.put((id*-1, str(id)))
        id_set.add(str(id))
    liansai_name = '英超'
    print(pq)
    indx = 0
    while not pq.empty():
        str_id = pq.get()[1]
        url = 'http://odds.500.com/fenxi/shuju-'+str_id+'.shtml'
        shuju_html = get_html(url)
        get_data_set(shuju_html, str_id)
        reg = liansai_name + '</a></td><td>\d*-\d*-\d*</td><td class="dz"><a href="./shuju-([0-9]\d*).shtml'
        page = re.compile(reg)
        rs = re.findall(page,shuju_html)
        for shuju_id in rs:
            if shuju_id not in id_set:
                id_set.add(shuju_id)
                pq.put((int(shuju_id)*-1, shuju_id))
                #print('append ', shuju_id)
        indx += 1

bfs_grab()
```
